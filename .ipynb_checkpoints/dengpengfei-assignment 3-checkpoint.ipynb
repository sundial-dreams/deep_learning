## Code Implementation (20 marks)

```python
import random
import string
import unicodedata
from abc import ABC

import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from matplotlib import ticker

languages = ["af", "cn", "de", "fi", "fr", "in", "ir", "pk", "za"]
n_languages = len(languages)

all_letters = string.ascii_letters + " .,;'"
n_letters = len(all_letters)

def unicodeToAscii(s):
    return "".join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn' and c in all_letters)


def getWords(baseDir, lang, train=True):
    filename = "{}/{}/{}.txt".format(baseDir, "train" if train else "val", lang)
    with open(filename, encoding='utf-8') as f:
        lines = f.read().strip().split('\n')
        return [unicodeToAscii(line) for line in lines]


def getLabels(lang, length):
    return [languages.index(lang) for _ in range(length)]


def readData(baseDir, train=True):
    x, y = [], []
    for lang in languages:
        x.append(getWords(baseDir, lang, train))
        y.append(lang)
    return x, y


def letter_to_index(letter):
    return all_letters.find(letter)


def line_to_tensor(line):
    tensor = torch.zeros(len(line), 1, n_letters)
    for li, letter in enumerate(line):
        tensor[li][0][letter_to_index(letter)] = 1
    return tensor

def category_from_output(output):
    top_n, top_i = output.topk(1)
    category_i = top_i[0].item()
    return languages[category_i], category_i


def randomChoice(l):
    i = random.randint(0, len(l) - 1)
    return l[i], i

def random_train_pair(X, y):
    y1, i = randomChoice(y)
    x1, _ = randomChoice(X[i])
    y1_tensor = torch.tensor([languages.index(y1)], dtype=torch.long)
    x1_tensor = line_to_tensor(x1)
    return y1, y1_tensor, x1, x1_tensor


# 计算准确度
def calculateAccuracy(model, X, y):
    acc = 0
    size = 0
    for i in range(len(X)):
        size += len(X[i])
        for j in range(len(X[i])):
            x_tensor = line_to_tensor(X[i][j])
            output, hidden = None, model.init_hidden()
            for k in range(x_tensor.size()[0]):
                output, hidden = model(x_tensor[k], hidden)
            guess, guess_i = category_from_output(output)
            acc += (1 if guess == y[i] else 0)
    return acc / size


def trainOneEpoch(model, criterion, optimizer, X, y):
    hidden = model.init_hidden()
    optimizer.zero_grad()
    output = None
    for i in range(X.size()[0]):
        output, hidden = model(X[i], hidden)

    loss = criterion(output, y)
    loss.backward()
    optimizer.step()
    return output, loss.item()


learning_rate = 0.0005  # 学习率
epoch = 100000  # 训练次数
n_hidden = 64  # 隐层数目
print_every = 500
plot_every = 1000

# 调参数
def train(lr, hidden_size=128, epoch=50000):
    model, all_losses, all_acc, val_losses, val_acc = run(lr, hidden_size, epoch)
    return all_losses, all_acc, val_losses, val_acc


def run(learning_rate=0.0005, hidden_size=64, epoch=100000):
    X, Y = readData("./datasets")
    X_val, Y_val = readData("./datasets", train=False)
    rnn = CharRNNClassify(n_letters, hidden_size, n_languages)
    current_loss = 0
    val_current_loss = 0
    current_acc = 0
    val_current_acc = 0
    all_losses = []
    all_acc = []
    val_losses = []
    val_acc = []

    criterion = nn.NLLLoss()
    optimizer = torch.optim.Adam(params=rnn.parameters(), lr=learning_rate)

    for e in range(1, epoch + 1):
        y, y_tensor, x, x_tensor = random_train_pair(X, Y)
        output, loss = trainOneEpoch(rnn, criterion, optimizer, x_tensor, y_tensor)
        current_loss += min(loss, 3)  # 防止loss过大
        guess, guess_i = category_from_output(output)
        current_acc += (1 if guess == y else 0)

        # 验证集
        y_val, y_val_tensor, x_val, x_val_tensor = random_train_pair(X_val, Y_val)
        output_val, hidden = None, rnn.init_hidden()

        for i in range(x_val_tensor.size()[0]):
            output_val, hidden = rnn(x_val_tensor[i], hidden)

        val_loss = criterion(output_val, y_val_tensor)
        val_current_loss += min(val_loss.item(), 3) # 防止loss过大
        guess_val, guess_val_i = category_from_output(output_val)

        val_current_acc += (1 if guess_val == y_val else 0)

#         if e % print_every == 0:
#             guess, guess_i = category_from_output(output)
#             correct = '✓' if guess == y else '✗ (%s)' % y
#             print('%d %.4f %f %s / %s' % (e, e / epoch * 100, loss, guess, correct))

        if e % plot_every == 0:
            all_losses.append(current_loss / plot_every)
            current_loss = 0
            all_acc.append(current_acc / plot_every)
            current_acc = 0
            # 验证集
            val_losses.append(val_current_loss / plot_every)
            val_current_loss = 0
            val_acc.append(val_current_acc / plot_every)
            val_current_acc = 0

    torch.save(rnn, "./model_classify.pkl")  # 保存模型
    return rnn, all_losses, all_acc, val_losses, val_acc
    # # 测试集准确度
    # acc = calculateAccuracy(rnn, X, Y)
    # print(acc)
    # # 验证集准确度
    # acc_val = calculateAccuracy(rnn, X_val, Y_val)
    # print(acc_val)
    # torch.save(rnn, "./model_classify.pkl")  # 保存模型
    #
    # # 训练集误差
    # plt.figure()
    # plt.title("loss and accuracy on train set")
    # plt.plot(all_losses, label="loss")
    # plt.plot(all_acc, label="accuracy")
    # plt.xlabel("epoch / 1000")
    # plt.ylabel("loss and accuracy")
    # plt.legend()
    # plt.show()
    #
    # # 验证集误差
    # plt.figure()
    # plt.title("loss and accuracy on validation set")
    # plt.plot(val_losses, label="validation loss")
    # plt.plot(val_acc, label="validation accuracy")
    # plt.xlabel("epoch / 1000")
    # plt.ylabel("loss and accuracy")
    # plt.legend()
    # plt.show()
    #
    # # 验证集混淆矩阵
    # confusion = evaluate(rnn, X_val, Y_val)
    # fig = plt.figure()
    # plt.title("confusion matrix")
    # ax = fig.add_subplot(111)
    # cax = ax.matshow(confusion.numpy())
    # fig.colorbar(cax)
    #
    # ax.set_xticklabels([''] + languages, rotation=90)
    # ax.set_yticklabels([''] + languages)
    #
    # ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    # ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    # plt.show()

    # model = torch.load("./mode_classify.pkl")


# 返回混淆矩阵
def evaluate(model, X, Y):
    confusion = torch.zeros(n_languages, n_languages)
    n_confusion = 10000
    for i in range(n_confusion):
        y, y_tensor, x, x_tensor = random_train_pair(X, Y)

        output, hidden = None, model.init_hidden()
        for j in range(x_tensor.size()[0]):
            output, hidden = model(x_tensor[j], hidden)
        guess, guess_i = category_from_output(output)

        category_i = languages.index(y)
        confusion[category_i][guess_i] += 1

    for i in range(n_languages):
        confusion[i] = confusion[i] / confusion[i].sum()

    return confusion
```

```python
import torch.nn as nn
from torch.autograd import Variable
import torch
import torch.nn.functional as F
from abc import ABC

# RNN model
class CharRNNClassify(nn.Module, ABC):
    def __init__(self, input_size, hidden_size, output_size):
        super(CharRNNClassify, self).__init__()
        self.hidden_size = hidden_size
        self.i2h = nn.Linear(input_size + hidden_size, hidden_size)
        self.i2o = nn.Linear(input_size + hidden_size, output_size)
        self.softmax = nn.LogSoftmax(dim=1)

    def forward(self, input, hidden=None):
        combined = torch.cat((input, hidden), 1)
        hidden = self.i2h(combined)
        output = self.i2o(combined)
        output = self.softmax(output)
        return output, hidden

    def init_hidden(self):
        return torch.zeros(1, self.hidden_size)

```

3. Experiment with the learning rate. You can try a few different learning rates and observe how this affects the loss. Another common practice is to drop the learning rate when the loss has plateaued. Use plots to explain your experiments and their effects on the loss (20 marks).

```python
'''
    学习率选择[0.001, 0.0005, 0.0001, 0.00005, 0.00001]这几组，可以发现训练轮数为50000轮时，学习率较好的为0.0005，0.001。
'''
def with_learning_rate(learning_rates=None):
    if learning_rates is None:
        learning_rates = [0.001, 0.0005, 0.0001, 0.00005, 0.00001]
    loss_list = []
    acc_list = []
    val_loss_list = []
    val_acc_list = []
    for lr in learning_rates:
        loss, acc, val_loss, val_acc = train(lr)
        loss_list.append(loss)
        acc_list.append(acc)
        val_loss_list.append(val_loss)
        val_acc_list.append(val_acc)

    plt.figure()
    for i in range(len(loss_list)):
        plt.plot(loss_list[i], label="learning_rate={}".format(learning_rates[i]))
    plt.title("loss with difference learning rate (hidden size = 128)")
    plt.ylabel("loss")
    plt.xlabel("epoch / 1000")
    plt.legend()

    plt.figure()
    for i in range(len(acc_list)):
        plt.plot(acc_list[i], label="learning_rate={}".format(learning_rates[i]))
    plt.title("accuracy with difference learning rate (hidden size = 128)")
    plt.ylabel("accuracy")
    plt.xlabel("epoch / 1000")
    plt.legend()
    plt.show()

    plt.figure()
    for i in range(len(val_acc_list)):
        plt.plot(val_acc_list[i], label="learning_rate={}".format(learning_rates[i]))
    plt.title("validation accuracy with difference learning rate (hidden size = 128)")
    plt.ylabel("validation accuracy")
    plt.xlabel("epoch / 1000")
    plt.legend()
    plt.show()

    plt.figure()
    for i in range(len(val_loss_list)):
        plt.plot(val_loss_list[i], label="learning_rate={}".format(learning_rates[i]))
    plt.title("validation loss with difference learning rate (hidden size = 128)")
    plt.ylabel("validation loss")
    plt.xlabel("epoch / 1000")
    plt.legend()
    plt.show()

with_learning_rate()
```
