Now that you know what DenseNet is all about, let's compare it to VGG.
Both VGG and DenseNet papers describe several variations of their models that differ by their depth.
For example, VGG16 and VGG19, DenseNet-121 and DenseNet-169 are four examples from these papers.

Aside from difference in network depth, how is the architecture of DenseNet different from that of VGG? Please enter your answer in the next cell (approximately 100-200 words, both English and Chinese are acceptable).

VGGNet的概念是更深的网络和更小的filters，VGGNet将AlexNet的层数进行增加，拥有16到19层的变体模型（VGG16、VGG19）。这些模型始终保持非常小的filters，并具有3\*3的卷积，基本上是最小的conv filter size。
DenseNet由Dense块组成，这些块中，各个层紧密连接在一起，每层都从先前层的输出特征映射中获取输入。残差具有极高的重用性并且产生了深度监督，每一层都从前一层接收到更多的监督。DenseNet对宽度和深度的需求很低，随着层之间的紧密联系，学习到的特征几乎没有冗余，层的输出特征映射数定义为增长率。最终，增长率控制了每一层向全局贡献了多少新信息。

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical

num_classes = 10

# load the CIFAR10 data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# input image dimensions
input_shape = x_train.shape[1:]

# mormalize data
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print('y_train shape:', y_train.shape)

# convert class vectors to binary class matrices.
# 可能会导致维数不一致的情况
# y_train = to_categorical(y_train, num_classes)
# y_test = to_categorical(y_test, num_classes)
```

## 3. Implement your models (30 marks)

In this task, you are required to implement DenseNet-121 and DenseNet-169 as depicted in the original paper.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from typing import List
num_classes = 10
EPSILON = 1.001e-5

def conv_block(x: tf.Tensor, growth_rate) -> tf.Tensor:
    x1 = layers.BatchNormalization(axis=3, epsilon=EPSILON)(x)
    x1 = layers.Activation('relu')(x1)
    x1 = layers.Conv2D(4 * growth_rate, 1, use_bias=False)(x1)
    x1 = layers.BatchNormalization(axis=3, epsilon=EPSILON)(x1)
    x1 = layers.Activation('relu')(x1)
    x1 = layers.Conv2D(growth_rate, 3, padding="same", use_bias=False)(x1)
    x = layers.Concatenate(axis=3)([x, x1])
    return x


def dense_block(x: tf.Tensor, blocks_len: int) -> tf.Tensor:
    for i in range(blocks_len):
        x = conv_block(x, 32)
    return x


def transition_block(x: tf.Tensor, reduction: float) -> tf.Tensor:
    x = layers.BatchNormalization(axis=3, epsilon=EPSILON)(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(int(keras.backend.int_shape(x)[3] * reduction), 1, use_bias=False)(x)
    x = layers.AveragePooling2D(2, strides=2)(x)
    return x

# 实现DenseNet 根据blocks不同而产生不同的DenseNet（DenseNet-161、DenseNet-121）
def DenseNet(blocks: List, shape: tuple) -> keras.Model:
    img_input = layers.Input(shape=shape)
    x = layers.ZeroPadding2D(padding=((3, 3), (3, 3)))(img_input)
    x = layers.Conv2D(64, 7, strides=2, use_bias=False)(x)
    x = layers.BatchNormalization(axis=3, epsilon=EPSILON)(x)
    x = layers.Activation('relu')(x)
    x = layers.ZeroPadding2D(padding=((1, 1), (1, 1)))(x)
    x = layers.MaxPooling2D(3, strides=2)(x)

    x = dense_block(x, blocks[0])
    x = transition_block(x, 0.5)
    x = dense_block(x, blocks[1])
    x = transition_block(x, 0.5)
    x = dense_block(x, blocks[2])
    x = transition_block(x, 0.5)
    x = dense_block(x, blocks[3])

    x = layers.BatchNormalization(axis=3, epsilon=EPSILON)(x)
    x = layers.Activation('relu')(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(10, activation='softmax')(x)

    return keras.Model(img_input, x)

# DenseNet-121
def DenseNet121(shape: tuple) -> keras.Model:
    return DenseNet([6, 12, 24, 16], shape)
```

```python
# DenseNet-169
def DenseNet169(shape: tuple) -> keras.Model:
    return DenseNet([6, 12, 32, 32], shape)
```

## 4. Train and evaluate your models. (25 marks)

### 4.1 Train your models. (20 marks)

```python
EPOCHS = 10 # 光十次mac就要跑一个多小时，没办法，实验室过于贫穷买不起卡（T_T）
def train_model(model: keras.Model) -> None:
    # 模型编译
    model.compile(
        optimizer=keras.optimizers.Adam(),
        loss=keras.losses.SparseCategoricalCrossentropy(),
        metrics=['accuracy']
    )

    # 训练模型
    history = model.fit(x_train, y_train, epochs=EPOCHS, validation_data=(x_test, y_test))

    # 绘制准确度与误差曲线
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    plt.figure(figsize=(8, 8))
    plt.subplot(1, 2, 1)
    plt.plot(range(EPOCHS), acc, label="training accuracy")
    plt.plot(range(EPOCHS), val_acc, label="validation accuracy")
    plt.legend()
    plt.title("training and validation accuracy")

    plt.subplot(1, 2, 2)
    plt.plot(range(EPOCHS), loss, label="training loss")
    plt.plot(range(EPOCHS), val_loss, label="validation loss")
    plt.legend()
    plt.title("training and validation loss")
    plt.show()
    
    print(
        "training accuracy: {}, training loss: {}, validation accuracy: {}, validation loss: {}"
          .format(acc[-1], loss[-1], val_acc[-1], val_loss[-1])
    )
    

# DenseNet-169    
model_169 = DenseNet169(shape=input_shape)
train_model(model_169)

# DenseNet-121
model_121 = DenseNet121(shape=input_shape)
train_model(model_121)
```

### 4.2 Test your models. (5 marks)

```python
def evaluate_model(model: keras.Model) -> None:
    # 评估模型
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)

    print("test loss: {}, test accuracy: {}".format(test_loss, test_acc))

# 测试模型    
evaluate_model(model_169)
evaluate_model(model_121)    
```

## 5. Load the pre-trained models from Keras and evaluate them. (15 marks)

```python
from tensorflow.keras.applications.densenet import DenseNet121 as DenseNet_121, DenseNet169 as DenseNet_169
from tensorflow.keras.applications import imagenet_utils

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

def evaluate_pre_model(model: keras.Model) -> tuple:
    predictions = imagenet_utils.decode_predictions(model.predict(x_test))
    total_acc = 0
    for i in range(len(predictions)):
        max_p, label = -1, ""
        for j in range(len(predictions[i])):
            if max_p < predictions[i][j][2]:
                max_p = predictions[i][j][2]
                label = predictions[i][j][1]
            
        if label.find(class_names[y_test[i][0]]) != -1:
            total_acc += 1

    total_acc = total_acc / len(y_test)
    total_loss = 1 - total_acc
    return total_acc, total_loss


x_test = tf.image.resize(x_test, (224, 224)) # 重设图片大小
model121 = DenseNet_121(input_shape=x_test[0].shape)
print("accuracy: {}, loss: {}".format(*evaluate_pre_model(model121)))

model169 = DenseNet_169(input_shape=x_test[0].shape)
print("accuracy: {}, loss: {}".format(*evaluate_pre_model(model169)))
```
